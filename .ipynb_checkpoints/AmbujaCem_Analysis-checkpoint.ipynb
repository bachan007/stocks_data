```python
import yfinance as yf
from datetime import datetime as dt
from dateutil import parser
import datetime
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
def date_format(date_val):
    date_val = parser.parse(date_val,dayfirst=True)
    formatted_date = date_val.strftime("%Y-%m-%d")
    if formatted_date:
        return formatted_date
    else:
        print("Enter a valid Date")
```

```python
start_date = date_format(input('Enter the Start date : \n'))
start_date
```

```python
end_date = input('Enter the End date : \n')
end_date
```

```python
if not end_date:
    end_date = dt.now().strftime('%Y-%m-%d')
    print(end_date)
else:
    end_date = date_format(end_date)
```

```python
# ticker = input("Enter the Stock Symbol : \n")
ticker = "AMBUJACEM.NS"
```

```python
data = yf.download(ticker,start_date,end_date)
data.sort_index(ascending=False, inplace=True)
# print(f"Here are the latest 5 records : \n {data.head()}")
```

```python
plt.figure(figsize=(12,7))
plt.plot(data.index,data['Adj Close'])
plt.title(f"{ticker} average price movement from {start_date} to {end_date}")
plt.xlabel('date')
plt.ylabel('price in INR')
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.plot(data.index,data['Open'])
plt.title('Open vs Date')

plt.subplot(1,2,2)
plt.plot(data.index,data['Close'])
plt.title('Close vs Date')

plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.plot(data.index,data.Open, label='Open')
plt.plot(data.index,data.Close, label='Close')
plt.title('Open vs Close')
plt.legend(['Open','Close'])
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.plot(data.index,data.High,color='Green')
plt.plot(data.index,data.Low, color='Red')
plt.legend(['High','Low'])
plt.title('High vs Low')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.plot(data.index,data.High-data.Low)
plt.title('(High - Low) Movement')
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.plot(data.index,data.Open-data.Close, color='Brown')
plt.title('(Open - Close) movement')
plt.grid()
plt.show()
```
