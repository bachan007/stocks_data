```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
from dateutil.relativedelta import relativedelta
import matplotlib.pyplot as plt
import mplfinance as mpf

import warnings
warnings.filterwarnings('ignore')
```

```python
end_date = dt.datetime.now()
start_date = (end_date-relativedelta(years=5)).strftime('%Y-%m-%d') # past five years performance
end_date = end_date.strftime('%Y-%m-%d')
```

```python
company = 'TECHM.NS'
```

```python
# Daywise data of the company
df_daily = yf.download(company,start_date,end_date)
```

```python
mpf.plot(df_daily.tail(100),type='candle', volume=True)
plt.show()
```
